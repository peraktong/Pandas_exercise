```python
# massive replacement

# use lambda to mass re-name columns:
# Create an example dataframe
data = {'Commander': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'Date': ['2012, 02, 08', '2012, 02, 08', '2012, 02, 08', '2012, 02, 08', '2012, 02, 08'], 
        'Score': [4, 24, 31, 2, 3]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
df.columns = ['Leader', 'Time', 'Score']


df.rename(columns={x:"i"+x for x in ['Leader', 'Time', 'Score']}, inplace=True)
df
```

```python
### similarly: If we want to massive re-name index, we can do this:
# df.index if the index
df.rename({x:"i"+x for x in df.index}, inplace=True)
df
```
